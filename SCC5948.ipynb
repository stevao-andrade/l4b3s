```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.

#How to read the files of the competition
#from subprocess import check_output
#print(check_output(["ls", "../input"]).decode("utf8"))

#read the files
test   = pd.read_csv('/home/stevao/Testes/inputs/test.csv')
train  = pd.read_csv('/home/stevao/Testes/inputs/train.csv')
sample = pd.read_csv('/home/stevao/Testes/inputs/sample_submission.csv')


#remove the ID column in train  and test
test  = test.drop(['ID'],axis=1)
train = train.drop(['ID'],axis=1)


#define the target. just get the information of the column target on train file
target = train.target


#first 5 lines
test.head()
```

```python
#create crossvalidation dataset using just the train.csv
cross_train = train.sample(frac = 0.7, random_state = 1)       #split it in 70% of train
cross_test  = train.loc[~train.index.isin(cross_train.index)]  #and 30%


#isolate the target from cross_train and cross_test datasets
cross_train_target = cross_train.target
cross_test_target  = cross_test.target

#delete the target information from the tables
cross_train = cross_train.drop(['target'],axis=1)
cross_test  = cross_test.drop(['target'],axis=1)


#categorical atributes
cat_cols = ['v3','v22','v24','v30','v31','v47','v52','v56','v66','v71','v74','v75','v79','v91','v107','v110','v112','v113','v125']


#handling categorical columns.. 

#converting into numerical
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()

for cat in cat_cols:
    
    cross_train[cat] = encoder.fit_transform(cross_train[cat])
    cross_test[cat]  = encoder.fit_transform(cross_test[cat])

    
#drop NaNs just to test
cross_train.fillna(0.0, inplace = True) 
cross_test.fillna(0.0 , inplace = True)
cross_train.head()
```

```python
#import KNN algorithm from scikitlearn package
from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier()

#fit the model to knn algorithm
knn.fit(cross_train,cross_train_target)

knn.predict(cross_test)

#check the score
print knn.score(cross_test, cross_test_target)


#decision tree
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(cross_train,cross_train_target)
tree.predict(cross_test)
print tree.score(cross_test, cross_test_target)


#naive bayes
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(cross_train,cross_train_target)
gnb.predict(cross_test)
print gnb.score(cross_test, cross_test_target)



#random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(cross_train,cross_train_target)
rf.predict(cross_test)
print rf.score(cross_test, cross_test_target)


knn.predict_proba(cross_test)
```
